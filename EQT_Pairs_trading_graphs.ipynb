```python
import pandas as pd
import matplotlib.pyplot as plt

financials_data_1994_24 = pd.read_csv('SP500_Financials_return_1994_24.csv')
healthcare_data_1994_24 = pd.read_csv('SP500_Heathcare_return_1994_24.csv')
it_data_1994_24 = pd.read_csv('SP500_IT_return_1994_24.csv')


financials_data_1994_24['Date'] = pd.to_datetime(financials_data_1994_24['Date'])
healthcare_data_1994_24['Date'] = pd.to_datetime(healthcare_data_1994_24['Date'])
it_data_1994_24['Date'] = pd.to_datetime(it_data_1994_24['Date'])


plt.figure(figsize=(8, 5))

plt.plot(financials_data_1994_24['Date'], financials_data_1994_24['Cumulative Profit'], label='Financials', color='green')
plt.plot(healthcare_data_1994_24['Date'], healthcare_data_1994_24['Cumulative Profit'], label='Healthcare', color='red')
plt.plot(it_data_1994_24['Date'], it_data_1994_24['Cumulative Profit'], label='Information Technology', color='blue')

plt.title('Cumulative Profit (1994-2024)')
plt.xlabel('Year')
plt.ylabel('Cumulative Profit')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
mr_data_path = 'Nikkei_MR_ranked_alpha0.1.csv'
nk225_data_path = 'nk225_adjclose_k1return_2004_24.csv'

mr_data = pd.read_csv(mr_data_path, parse_dates=['Date'])
nk225_data = pd.read_csv(nk225_data_path, parse_dates=['Date'])

# Synchronize dates
common_start = max(mr_data['Date'].min(), nk225_data['Date'].min())
common_end = min(mr_data['Date'].max(), nk225_data['Date'].max())
mr_data_sync = mr_data[(mr_data['Date'] >= common_start) & (mr_data['Date'] <= common_end)]
nk225_data_sync = nk225_data[(nk225_data['Date'] >= common_start) & (nk225_data['Date'] <= common_end)]

# Define rolling Sharpe ratio function
def rolling_sharpe(returns, window):
    mean = returns.rolling(window=window).mean()*252
    std = returns.rolling(window=window).std()*252**0.5
    sharpe = mean / std
    return sharpe

# Calculate rolling Sharpe ratio with a window of 1260 trading days
window_size = 1260
mr_sharpe = rolling_sharpe(mr_data_sync['Profit'], window_size)
nk225_sharpe = rolling_sharpe(nk225_data_sync['Profit'], window_size)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(nk225_data_sync['Date'], nk225_sharpe, label='Pairs Trading Rolling Sharpe Ratio', color='orange')
plt.plot(mr_data_sync['Date'], mr_sharpe, label='Mean Reversion Rolling Sharpe Ratio', color='blue')
plt.title('NK225 5-Year Rolling Sharpe Ratio Comparison')
plt.xlabel('Date')
plt.ylabel('Sharpe Ratio')
plt.legend()
plt.grid(True)
plt.show()
```
